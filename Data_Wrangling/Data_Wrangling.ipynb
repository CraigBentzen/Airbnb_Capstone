```python
#dropping the column from the dataframe if the amenities occurs less than 15 times.  I'll need to revisit this number when I include the entire dataset
for x in df_amenities.columns:
    if (df_amenities[x] == 1).sum() < 15: 
        df_amenities.drop(x,axis=1, inplace = True)
```
